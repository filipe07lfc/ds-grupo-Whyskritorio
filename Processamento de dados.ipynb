```python
#Substitui os NaN
db = dados_brutos.fillna(0)
display(db)
```

```python
# Inflacao de 5.86% anual (2012 -> 2019)
# Multiplicar por 1.0586(Para cada ano)
db[['income', 'income_work', 'income_rent', 'income_capital']] = db[['income', 'income_work', 'income_rent', 'income_capital']] * ((1.0586)**7)
display(db)
```
